Complete all questions using Pandas.

## Q1. Import pandas and load the dataset.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("movies_dataset.csv")
```

```python
df[df.duplicated()]
```

## Q10. Remove duplicate rows.

```python
df.drop_duplicates(inplace=True)
```

## Q11. Remove extra spaces from Title.

```python
df['Title'] = df['Title'].str.strip()
```

## Q12. Convert Title to Proper Case.

```python
df['Title'] = df['Title'].str.title()
```

## Q13. Correct Genre spellings (Acton→Action, Scifi→Sci-Fi, Romnce→Romance).

```python
df['Genre'] = df['Genre'].replace({
    'Acton': 'Action',
    'Scifi': 'Sci-Fi',
    'Romnce': 'Romance'
})
```

## Q14. Convert Country to upper case.

```python
df['Country'] = df['Country'].str.upper()
```

## Q15. Convert Language to lower case.

```python
df['Language'] = df['Language'].str.lower()
```

## Q16. Convert Rating to float.

```python
df['Rating'] = df['Rating'].astype(float)
```

## Q17. Remove 'min' from Duration and convert to integer.

```python
df['Duration'] = df['Duration'].str.replace('min', '', regex=False).str.strip().astype(int)
```

## Q18. Remove '$' and 'M' from Budget.

```python
df['Budget'] = df['Budget'].str.replace('$', '', regex=False)\
                           .str.replace('M', '', regex=False)\
                           .astype(float)
```

## Q19. Remove commas from Revenue and convert numeric.

```python
df['Revenue'] = pd.to_numeric(
    df['Revenue'].str.replace(',', '', regex=False),
    errors='coerce'
)
```

## Q20. Convert Release_Year to integer.

```python
df['Release_Year'] = df['Release_Year'].astype(int)
```

## Q21. Convert Added_Date to datetime.

```python
df['Added_Date'] = pd.to_datetime(df['Added_Date'])
```

## Q22. Delete rows where Title is missing.

```python
df.dropna(subset=['Title'], inplace=True)
```

## Q23. Fill missing Rating with mean.

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

## Q24. Fill missing Country with mode.

```python
df['Country'] = df['Country'].fillna(df['Country'].mode()[0])
```

## Q25. Fill missing Budget with median.

```python
df['Budget'] = df['Budget'].fillna(df['Budget'].median())
```

## Q26. Check if any null values remain.

```python
df.isnull().sum()
```

## Q27. Save cleaned dataset as clean_movies_dataset.csv.

```python
df.to_csv('clean_movies_dataset.csv', index=False)
print("File saved successfully as clean_movies_dataset.csv")
```
