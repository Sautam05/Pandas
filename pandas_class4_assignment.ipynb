
# Q9. Group By City

## Problem Statement
Find average marks city-wise.

## Sample Input
```python
City and Marks
```

## Sample Output
```python
Grouped result
```


````python
# Write your code here
import pandas as pd
data = {
    "Name": ["Aman", "Riya", "Rahul", "Priya"],
    "City": ["Bhopal", "Indore", "Bhopal", "Indore"],
    "Marks": [80, 90, 70, 85]
}

df = pd.DataFrame(data)
print("Grouped result")
print(df.groupby("City")["Marks"].mean())
````


# Q10. Multiple Conditions

## Problem Statement
Display students with Marks > 70 and Age > 20.

## Sample Input
```python
Marks and Age
```

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
import pandas as pd
data = {
    'Name': ["Prem", "Krish", "Vikram", "Sautam"],
    'Marks': [78,85,83,89],
    'Age': [20, 18, 22, 21]
}
df = pd.DataFrame(data)
print(df[(df['Age'] > 20) & (df['Marks'] > 80)])
````
