## Q1. Find Missing Values

Input:
df = pd.DataFrame({
    "Name":["Aman","Riya",None],
    "Age":[21,np.nan,25]
})

Output:
    Name    Age
0  False  False
1  False   True
2   True  False

```python
# Write your code here
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Name": ["Aman", "Riya", None],
    "Age": [21, np.nan, 25]
})

print(df.isnull())
```

## Q2. Count Total Missing Values

Input:
df = pd.DataFrame({

    "Age":[20,None,25]
})

## Q3. Fill Missing Age with 0

Input:
df = pd.DataFrame({
    "Name":["Aman","Raj"],
    "Age":[20,None]
})

Output:
   Name   Age
0  Aman  20.0
1   Raj   0.0

## Q7. Replace Delhi with New Delhi

## Q13. Mini Data Cleaning Task

Input:
data = {
    "Name":["Aman","Raj","Aman",None],
    "Age":[20,None,20,25],
    "City":["Bhopal","Delhi","Bhopal","Indore"]
}

Task:
1. Fill missing Age with mean.
2. Remove duplicate rows.
3. Display final DataFrame.

```python
# Write your code here
data = {
    "Name": ["Aman", "Raj", "Aman", None],
    "Age": [20,None,20,25],
    "City": ["Bhopal", "Delhi", "Bhopal","Indore"]
}
b = pd.DataFrame(data)
b["Age"].fillna(b["Age"].mean(), inplace=True)
b.drop_duplicates(inplace=True)
print(b)
```
