# Q1. Display Unique Values

## Problem Statement
Create a DataFrame with a City column and display all unique cities.

## Sample Input
```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output
```python
['Bhopal' 'Indore' 'Delhi']
```


````python
# Write your code here
import pandas as pd
a = {
    'City': ["Bhopal", "Indore", "Bhopal", "Delhi"]
}
df = pd.DataFrame(a)
print(df['City'].unique())
````


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
# Write your code here
import pandas as pd
a = {
    'City': ["Bhopal", "Bhopal", "Delhi"],
    'Marks': [80,90,70]
}
b = pd.DataFrame(a)
print(b.groupby('City')['Marks'].mean())
````


# Q5. Sort Values


# Q8. Multiple Conditions

## Problem Statement
Display students whose marks are greater than 70 and age is greater than 20.

## Sample Output
```python
Filtered rows
```


````python
# Write your code here
import pandas as pd
data = {
    "Name": ["Aman", "Riya", "Rahul", "Sneha"],
    "Age": [20, 21, 19, 22],
    "Marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)
print(df[(df["Marks"] > 70) & (df["Age"] > 20)])
````


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000
IT      120000
```


````python
# Write your code here
import pandas as pd
data = {
    "Employee": ["Aman", "Riya", "Rahul", "Sneha"],
    "Department": ["HR", "IT", "IT", "HR"],
    "Salary": [20000, 50000, 70000, 30000]
}

df = pd.DataFrame(data)
print(df.groupby("Department")["Salary"].sum())
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
# Write your code here
import pandas as pd
data = {
    "Name": ["Aman", "Riya", "Rahul", "Sneha", "Karan"],
    "Age": [20, 21, 19, 22, 20],
    "City": ["Delhi", "Mumbai", "Delhi", "Bhopal", "Mumbai"],
    "Marks": [85, 90, 78, 88, 82]
}

df = pd.DataFrame(data)
print(df["City"].unique())

print(df["City"].value_counts())

print(df[df["Marks"] > 80])

print(df.sort_values(by="Marks", ascending=False))
````
