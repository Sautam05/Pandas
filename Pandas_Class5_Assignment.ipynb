## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
# Write your code here
import pandas as pd
a = {
    "Age": [25,18,30,22]
}
b = pd.DataFrame(a)
print(b.sort_values(by="Age"))
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
# Write your code here
import pandas as pd
data = {
    'Name': ["Sautam", "Akash", "Vishal", "Nitish", "Arti"],
    'Age': [20,19,25,30,21],
    'Score': [89,70,75,84,75],
    'Height': [5.7,5.4,5.2,5.3,5.5]
}
df = pd.DataFrame(data)
print(df.sort_values(by='Score', ascending=False))
print(df['Age'].mean())
print(df['Height'].median())
print(df['Score'].value_counts())
```
